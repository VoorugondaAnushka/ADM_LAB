```python
import pandas as pd
```

```python
df = pd.read_csv('/content/employee_data .csv')
print('Original dataset')
print(df)
```

```python
df.isna()
```

```python
df.isnull().sum()
```

```python
df['Age'].median()
```

```python
df['Salary'].mean()
```

```python
df['Job_Satisfaction'].mode()
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
```

```python
df['Job_Satisfaction'] = df['Job_Satisfaction'].fillna(df['Job_Satisfaction'].mode()[0])
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
numerical_features = ['Age', 'Salary', 'Job_Satisfaction']
# Assuming these are the numerical columns in your DataFrame 'df'
df[numerical_features] = scaler.fit_transform(df[numerical_features])

print("\nPreprocessed Dataset:")
print(df)
```
